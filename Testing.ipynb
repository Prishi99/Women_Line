```python
from google.colab import files
uploaded = files.upload()

# Load your dataset
df = pd.read_csv("annotated_prompts_with_intent.csv")

# Drop unused columns
df.drop(columns=['Language', 'Context', 'Tags'], inplace=True)

# Preview
df.head()
```

```python
# Check label distribution to understand imbalance
print("Intent distribution:\n", df['Intent'].value_counts())
print("\nSymptom distribution:\n", df['Symptom'].value_counts())
print("\nSentiment distribution:\n", df['Sentiment'].value_counts())
```

```python
# Seed prompts for each intent
intent_examples = {
    "mental_wellness": [
        "How to reduce anxiety during periods?",
        "Tips to improve mood swings?"
    ],
    "diet_advice": [
        "Best foods to eat during menstruation?",
        "Can spicy food worsen cramps?"
    ],
    "exercise_guidance": [
        "Can I do yoga during periods?",
        "What workouts are safe on my period?"
    ],
    "myth_busting": [
        "Can I wash my hair on first day of period?",
        "Is it true you can’t swim during periods?"
    ],
    "symptom_query": [
        "Why do I feel dizzy before my period?",
        "Are breast pains normal during PMS?"
    ]
}

# Prompt generation
def generate_prompts(template_list, label, count=50):
    prompts = []
    for _ in range(count):
        prompt = random.choice(template_list)
        prompts.append({
            "Prompt": prompt,
            "Intent": label,
            "Symptom": "cramps" if "pain" in prompt or "cramps" in prompt else "mood",
            "Sentiment": "negative" if "pain" in prompt or "anxiety" in prompt else "positive"
        })
    return prompts

# Combine all generated prompts
generated = []
for intent, templates in intent_examples.items():
    generated += generate_prompts(templates, intent)

df_gen = pd.DataFrame(generated)
```

```python
translator = Translator()

# Function to translate prompt
def translate_prompt(text, lang_code):
    try:
        return translator.translate(text, dest=lang_code).text
    except:
        return ""

# Apply translations
df_gen['Hindi Translation'] = df_gen['Prompt'].apply(lambda x: translate_prompt(x, 'hi'))
df_gen['Tamil Translation'] = df_gen['Prompt'].apply(lambda x: translate_prompt(x, 'ta'))
df_gen['Bengali Translation'] = df_gen['Prompt'].apply(lambda x: translate_prompt(x, 'bn'))
df_gen['Marathi Translation'] = df_gen['Prompt'].apply(lambda x: translate_prompt(x, 'mr'))
df_gen['Urdu Translation'] = df_gen['Prompt'].apply(lambda x: translate_prompt(x, 'ur'))
df_gen['Telugu Translation'] = df_gen['Prompt'].apply(lambda x: translate_prompt(x, 'te'))
df_gen['Code-Mixed (Hinglish)'] = df_gen['Prompt'].apply(lambda x: "Yaar, " + x.replace("?", " kya?"))
df_gen['valid_prompt'] = True  # You can add real validation logic later
```

```python
df['valid_prompt'] = True
df.fillna("", inplace=True)  # In case any translations are missing
```

```python
final_df = pd.concat([df, df_gen], ignore_index=True)

# Save for inspection if needed
final_df.to_csv("final_balanced_translated_dataset.csv", index=False, encoding="utf-8")
final_df.sample(5)
```

```python
# Use prompt text to predict intent
X = final_df['Prompt']
y = final_df['Intent']

# TF-IDF for better feature extraction
vectorizer = TfidfVectorizer()
X_vec = vectorizer.fit_transform(X)

# Split
X_train, X_test, y_train, y_test = train_test_split(X_vec, y, test_size=0.2, random_state=42)

# Train a stronger model
model = RandomForestClassifier(n_estimators=150, random_state=42)
model.fit(X_train, y_train)

# Evaluate
y_pred = model.predict(X_test)
print("✅ Accuracy:", accuracy_score(y_test, y_pred))
print("\n📊 Classification Report:\n", classification_report(y_test, y_pred))
```

```python
from deep_translator import GoogleTranslator

def translate_all(text):
    langs = {
        'Hindi': 'hi',
        'Tamil': 'ta',
        'Bengali': 'bn',
        'Marathi': 'mr',
        'Urdu': 'ur',
        'Telugu': 'te'
    }
    translations = {}
    for lang, code in langs.items():
        try:
            translated = GoogleTranslator(source='auto', target=code).translate(text)
            translations[f"{lang} Translation"] = translated
        except Exception as e:
            translations[f"{lang} Translation"] = 'Translation Failed'
            print(f"Error in {lang}: {e}")
    return translations
```

```python
all_data = []
for intent, prompts in real_prompts.items():
    for p in prompts:
        row = {
            'Prompt': p,
            'Intent': intent,
            'Symptom': '',
            'Sentiment': 'negative',
            'valid_prompt': True
        }
        row.update(translate_all(p))
        row['Code-Mixed (Hinglish)'] = p + " kya?"
        all_data.append(row)

df_gen = pd.DataFrame(all_data)
```
